```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
#Top Video Game
headers = {'Accept-Language': 'en-US,en;q=0.8'}
url = 'https://www.imdb.com/search/title/?title_type=game&sort=user_rating,desc'
response = requests.get(url,headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
rank = []
game_title = []
year = []
link = []
certificate = []
genre = []
```

```python
for item in soup.select('div.lister-item.mode-advanced'):
    rank_class = item.find('span', class_='lister-item-index unbold text-primary')
    rank_info = rank_class.get_text(strip=True) if rank_class else ''
    rank_value = int(rank_info.split('.')[0]) if rank_info else None
    rank.append(rank_value)

    game_class= item.find('h3', class_='lister-item-header').find('a')
    game_text = game_class.get_text(strip=True) if game_class else ''
    game_title.append(game_text)

    link_class = game_class.get('href') if game_class else ''
    link_complete = f"https://www.imdb.com{link_class}?ref_=adv_li_tt"
    link.append(link_complete)

    year_class = item.find('span', class_='lister-item-year')
    year_info = year_class.get_text(strip=True).replace("(", "").replace(")", "") if year_class else ''
    year.append(year_info)

    certificate_class = item.find('span', class_='certificate')
    certificate_info = certificate_class.get_text(strip=True) if certificate_class else ''
    certificate.append(certificate_info)

    genre_class = item.find('span', class_='genre')
    genre_info = genre_class.get_text(strip=True).replace("\n", "").replace(" ", "") if genre_class else ''
    genre.append(genre_info)
```

```python
min_length = min(len(rank), len(game_title), len(link), len(year), len(certificate), len(genre))
```

```python
df = pd.DataFrame(
    {'Rank': rank[:min_length],
     'Game': game_title[:min_length],
     'Link': link[:min_length],
     'Year of Release': year[:min_length],
     'Rating/Certification': certificate[:min_length],
     'Genre': genre[:min_length],}
)
print(df.head())
```

```python
df.to_csv('gamesrating.csv', index=False)
pd.read_csv("gamesrating.csv", index_col=0)
```
